**üéØ Step 4: Feature Engineering**

**1. Handling Availability:**

```python
# Step 1: Handle availability column
def availability_to_numeric(availability):
    if 'left' in str(availability).lower():
        # Extract the number of items left
        return int(''.join(filter(str.isdigit, str(availability))))
    elif 'in stock' in str(availability).lower():
        return 1  # In stock (1 means available)
    elif 'cannot be shipped' in str(availability).lower():
        return 0  # Not available (0 means unavailable)
    else:
        return np.nan  # In case there are unexpected values

# Apply the function to 'availability' column
df['availability_numeric'] = df['availability'].apply(availability_to_numeric)

# Check the transformed data
df[['availability', 'availability_numeric']].head()
```

**2. Extract Numerical Data from `tech_item_weight`:**

```python
# Step: Improved extraction of weight from 'tech_item_weight'
def extract_weight(weight_str):
    if isinstance(weight_str, str):
        # Remove 'ounces', 'lbs', 'kg', and spaces, then split by semicolon
        weight_str = weight_str.replace('ounces', '').replace('lbs', '').replace('kg', '').replace(' ', '').replace(';', ',')

        # Split by commas if multiple weights exist
        weights = weight_str.split(',')

        # Try to convert each weight and calculate the average if there are multiple
        try:
            numeric_weights = [float(weight) for weight in weights if weight.replace('.', '', 1).isdigit()]
            if len(numeric_weights) > 0:
                return np.mean(numeric_weights)  # Take average of multiple weights
            return np.nan
        except ValueError:
            return np.nan
    return np.nan

# Apply the improved function to 'tech_item_weight'
df['tech_item_weight_numeric'] = df['tech_item_weight'].apply(extract_weight)

# Check the result
df[['tech_item_weight', 'tech_item_weight_numeric']].head()
```

```python
# Step: Improved extraction of dimensions from 'tech_product_dimensions'
def extract_dimensions(dim_str):
    if isinstance(dim_str, str):
        # Remove non-numeric characters (e.g., 'inches', 'cm', 'ounces')
        dim_str = dim_str.replace('inches', '').replace('cm', '').replace('ounces', '').replace(' ', '')

        # Split by 'x' for standard dimensions or by ';' for multiple values
        dims = dim_str.replace(';', 'x').split('x')

        # Convert each dimension to float and return average if multiple
        try:
            dims = [float(dim) for dim in dims if dim.replace('.', '', 1).isdigit()]
            if len(dims) == 3:
                return dims  # Return length, width, and height if all three are present
            elif len(dims) > 0:
                return [np.nan, np.nan, np.nan]  # Not enough dimensions
        except ValueError:
            return [np.nan, np.nan, np.nan]  # In case conversion fails
    return [np.nan, np.nan, np.nan]

# Apply to the 'tech_product_dimensions' column
df[['tech_length', 'tech_width', 'tech_height']] = df['tech_product_dimensions'].apply(lambda x: pd.Series(extract_dimensions(x)))

# Check the result
df[['tech_product_dimensions', 'tech_length', 'tech_width', 'tech_height']].head()
```
